## Transform

In this step, I prepared the data for analysis.

First, I converted the date column into datetime format and extracted year and month for time-based analysis.

I then created new features like revenue and average order value.

Finally, I standardized the data by cleaning column names and important text fields, removing extra spaces and converting them to lowercase to ensure consistency.

```python
print("*"*10,"Convert date column","*"*10)
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
print("*"*10,"Extract time features","*"*10)
df['YEAR'] = df['ORDERDATE'].dt.year
df['MONTH'] = df['ORDERDATE'].dt.month
```

```python
print("*"*10,"Create revenue column","*"*10)
df['REVENUE'] = df['SALES']
```

```python
print("*"*10,"Average order value","*"*10)
df['AVG_ORDER_VALUE'] = df['REVENUE'] / df['QUANTITYORDERED']
```

```python
df.head(5)
```

```python
print("*"*10,"Fixing the string of column names","*"*10)
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
print("*"*10,"Fixing the strign of some important column value","*"*10)
df['productline'] = df['productline'].str.strip().str.lower()
df['country'] = df['country'].str.strip().str.lower()
df['dealsize'] = df['dealsize'].str.strip().str.lower()
```

```python
df.head()
```

### Data Saving

After cleaning and transforming the data, I saved the final dataset for future use and reproducibility.

```python
print("*"*10,"Saving the clean data","*"*10)
df.to_csv("../data/cleaned/cleaned_sales.csv", index=False)
```

## Analyze

In this step, I explored the data to understand different aspects of the business.

I analyzed sales trends over time, checked which products generate the most revenue, and looked at performance across different countries and deal sizes.

I also analyzed customer behavior to understand repeat purchases.

This helps in identifying key drivers of revenue and overall business performance.

### Product Analysis

```python
print("*"*10,"Product Analysis","*"*10)

monthly_sales = df.groupby(['year','month'])['revenue'].sum().reset_index()

monthly_sales['growth_%'] = monthly_sales['revenue'].pct_change() * 100

monthly_sales.head()
```

### Country Analysis

```python
print("*"*10,"Country Analysis","*"*10)

country_sales = df.groupby('country')['revenue'].sum().sort_values(ascending=False)

country_sales.head()
```

### Deal Size Analysis

```python
print("*"*10,"Deal Size Analysis","*"*10)

deal_sales = df.groupby('dealsize')['revenue'].sum()

deal_sales
```

### Customer Analysis

```python
print("*"*10,"Customer Analysis","*"*10)

customer_orders = df.groupby('customername')['ordernumber'].nunique()

repeat_ratio = (customer_orders > 1).sum() / customer_orders.count() * 100

print("Repeat Ratio : ",repeat_ratio)
```

```python
product_sales = df.groupby('productline')['revenue'].sum().sort_values(ascending=False)

product_percent = (product_sales / product_sales.sum()) * 100

product_percent
```

<hr>

## Visualize

In this step, I created visualizations to better understand the patterns in the data.

Graphs help in identifying trends, comparing performance, and making the analysis easier to interpret.

I used line charts for sales trends and bar charts for comparing categories like product lines, countries, and deal sizes.

### Sales Trend

```python
plt.figure(figsize=(10,5))

plt.plot(monthly_sales.index, 
         monthly_sales['revenue'], 
         marker='o', 
         color='#6C5CE7',
         label='Revenue')

plt.title("Sales Trend")
plt.xlabel("Time", labelpad=15)
plt.ylabel("Revenue", labelpad=15)

plt.legend()

plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig("../outputs/sales_trend.png")

plt.show()
```

### Product Contribution

```python
plt.figure(figsize=(10,5))

product_percent.plot(kind='bar', label='Product %')

plt.title("Product Contribution")
plt.xlabel("Product Line", labelpad=15)
plt.ylabel("Percentage", labelpad=15)

plt.legend()

plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig("../outputs/product_sales.png")

plt.show()
```

### Country Sales

```python
plt.figure(figsize=(10,5))

country_sales.head(10).plot(kind='bar', label='Revenue')

plt.title("Top Countries")
plt.xlabel("Country", labelpad=15)
plt.ylabel("Revenue", labelpad=15)

plt.legend()

plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig("../outputs/country_sales.png")

plt.show()
```

### Deal Size

```python
plt.figure(figsize=(8,5))

deal_sales.plot(kind='bar', label='Revenue')

plt.title("Deal Size")
plt.xlabel("Deal Size")
plt.ylabel("Revenue")

plt.legend()

plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig("../outputs/deal_size.png")

plt.show()
```

## Insights
